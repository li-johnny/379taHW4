## Deliverable 4C

```python
FreB = pd.read_csv("InvoicePurchases12-31-16Sample.csv")
FreB.head(2)
```

```python
FreB = FreB.fillna(0)
```

```python
FreB_Group= FreB.groupby(['VendorNumber', 'VendorName'], as_index=False)[['Dollars', 'Quantity','Freight']].sum()
```

```python
Frei = FreB_Group[FreB_Group["Dollars"] >= 250000]
Frei.head(2)
```

```python
Frei.loc[:,'FrePerUnit'] = Frei['Freight'] / Frei['Quantity']
Frei.head()
```

```python
Frei['FreightRate(%)'] = (Frei["Freight"]/Frei['Dollars'])*100 #creates new column
```

```python
Frei.sort_values(by='FreightRate(%)',ascending=False).head(10)
#sorting out the highest 10
#sort by freight rate ascen
```

```python
Frei.sort_values(by='FreightRate(%)',ascending=True).head(5)
#sort out the lowest 5                                                    
```

## Deliverable 4D

```python
Freig = FreB[FreB["Freight"] > 100]
Freig.head(2)
```

```python
smallExpFre = Freig[Freig["Quantity"] <= 1000]
smallExpFre.head()
```

```python
smallExpFre= smallExpFre.groupby(['VendorNumber', 'VendorName'], as_index=False)[['Dollars', 'Quantity','Freight']].sum()
```

```python
# create FreightperUnit 
smallExpFre.loc[:,'FrePerUnit'] = smallExpFre['Freight'] / smallExpFre['Quantity']
smallExpFre.head()
```

```python
# create freight per $
smallExpFre['FreightRate(%)'] = (smallExpFre["Freight"]/smallExpFre['Dollars'])*100 #creates new column
```

```python
smallExpFre.sort_values(by="Freight", ascending=False).head(5)
```
