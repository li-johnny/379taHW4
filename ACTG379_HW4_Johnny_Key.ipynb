## B

```python
#Recognized on Invoice Date
invoices_2016ID = invoices[invoices["InvoiceDate"].str.startswith("2016")]

#Generates an overall list of said values in groupby and agg()
topVendAll = invoices_2016ID.groupby(["VendorNumber", "VendorName"]).agg({"Dollars": "sum", "Freight": "sum", "Quantity": "sum"}).reset_index()

#Generates a filtered list of just the requested values
topVendCharge = topVendAll[["VendorNumber", "VendorName", "Freight"]].sort_values("Freight", ascending=False).reset_index(drop=True)

topVendCharge.head(10)
```

```python
# #Generates new column to hold the rate
# topVendAll["Freight Per $ Purchased"] = (topVendAll["Freight"] / topVendAll["Dollars"]) 

# #Filters out all the rows not >=250k
# topVend250 = topVendAll[topVendAll["Dollars"] >= 250000].copy(deep=True) 

# topVend250 = topVend250[["VendorNumber", "VendorName", "Freight Per $ Purchased"]].sort_values("Freight Per $ Purchased", ascending=False).reset_index(drop=True)

# # topVend250["Freight Per $ Purchased"] = topVend250["Freight Per $ Purchased"].round(4).map("${:.4f}".format)
```

```python
#Filters out all the rows not >=250k
topVend250 = topVendAll[topVendAll["Dollars"] >= 250000].copy(deep=True) 

#Generates new column to hold the rate
topVend250["FreightRate$(%)"] = (topVend250["Freight"] / topVend250["Dollars"])*100
topVend250["FreightPerUnit"] = (topVend250["Freight"] / topVend250["Quantity"])

topVend250 = topVend250[["VendorNumber", "VendorName", "Freight", "Dollars", "Quantity", "FreightRate$(%)","FreightPerUnit"]].sort_values("FreightRate$(%)", ascending=False).reset_index(drop=True)
```

## D

```python
#Filter out from the 2016 dataframe to take requested values
d4Master = invoices_2016ID[(invoices_2016ID["Freight"] > 100) & (invoices_2016ID["Quantity"] <= 1000)].copy(deep=True)

# d4_d["Freight Per $ Purchased"] = (d4_d["Freight"] / d4_d["Dollars"])
d4_d = d4Master.groupby(["VendorNumber", "VendorName"]).agg({"Dollars": "sum", "Quantity": "sum", "Freight": "sum"}).reset_index().sort_values("Freight", ascending=False)

# d4_d = d4_d.groupby(["VendorNumber", "VendorName", ]).agg({"Dollars": "sum", "Freight": "sum", "Quantity": "sum", "Freight Per $ Purchased": "sum"}).reset_index()

# d4_d = d4_d.groupby(["VendorNumber", "VendorName",  "Freight Per $ Purchased"]).agg({"Dollars": "sum", "Freight": "sum", "Quantity": "sum"}).reset_index()
d4_d = d4_d[["VendorNumber", "VendorName", "Freight", "Dollars", "Quantity"]].reset_index(drop=True)
d4_d.head()
```

## E

# Deliverable 5


```python
mergeSales = sales.groupby("InventoryId").agg({"SalesQuantity": "sum", "SalesDollars": "sum"}).sort_values("SalesQuantity", ascending=False).reset_index()
```

```python
# Merge sales inventory on the inventory ID column
mergeSales = pd.merge(mergeSales, totalInv, how="outer", on=["InventoryId"])
```

```python
# Calculate purchases by subtracting the beginning inventory quantity from the ending inventory quantity, and then adding the sales quantity
mergeSales["Purchases"] = (mergeSales["Ending Quantity"] - mergeSales["Beginning Quantity"]) + mergeSales["SalesQuantity"]
mergeSales.fillna(0, inplace=True)
mergeSales[["SalesQuantity", "Beginning Quantity", "Ending Quantity", "Purchases"]] = mergeSales[["SalesQuantity", "Beginning Quantity", "Ending Quantity","Purchases"]].astype(int)
```

```python
mergeSales = mergeSales.sort_values("Purchases", ascending=False).reset_index(drop=True)
mergeSales = mergeSales[["InventoryId", "SalesQuantity", "Beginning Quantity", "Ending Quantity", "Purchases"]]
```
