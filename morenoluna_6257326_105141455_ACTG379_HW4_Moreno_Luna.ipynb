## Deliverable 4: Purchases 

#### (A) Top 10 Vendors

```python
purchases = pd.read_csv("PurchasesFINAL12312016.csv")
purchases.head(2)
```

```python
vendors = purchases[['VendorName','Dollars']].groupby('VendorName').sum()
vendors.head(2)
```

```python
vendors.sort_values(by='Dollars',ascending=False).head(10)
```

#### (B) Freight Top 10 Vendors 

```python
freight = pd.read_csv("InvoicePurchases12312016.csv")
freight.head(2)
```

```python
freight2 = freight[['VendorName','Freight']].groupby('VendorName').sum()
freight2.head(2)
```

```python
freight2.sort_values(by='Freight',ascending=False).head(10)
```

#### (C) Freight Tables 

```python
freight_purchases=pd.merge(left=vendors,right=freight2,on='VendorName',how='outer')
freight_purchases.head(2)
```

```python
at_least= freight_purchases[freight_purchases['Dollars']>=250000]
at_least.head()
```

```python
highest= freight_purchases.groupby('VendorName').sum()
highest.head(2)
```

```python
highest.sort_values(by='VendorName',ascending=False).head(10)
```

```python
lowest= freight_purchases.groupby('VendorName').sum()
lowest.head(2)
```

```python
lowest.sort_values(by='VendorName',ascending=True).head(5)
```
