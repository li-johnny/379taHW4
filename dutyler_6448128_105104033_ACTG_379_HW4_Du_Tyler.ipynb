# Deliverable 5

```python
Sales = pd.read_csv("C:/Users/Tyler Du/Downloads/SimplifiedData/SimplifiedData/HW4 Sampled Data/SalesFINAL12312016SAMPLE.csv")
Sales.head()
```

```python
Purchases = pd.read_csv("C:/Users/Tyler Du/Downloads/SimplifiedData/SimplifiedData/HW4 Sampled Data/PurchasesFINAL12312016SAMPLE.csv")
Purchases.head()
```

```python
Deliverable5_1 = pd.merge(left = Sales[['InventoryId', 'Store', 'Brand', 'SalesDollars', 'SalesPrice']], right = Purchases[['InventoryId', 'PurchasePrice', 'Dollars']], how = 'inner', on = ['InventoryId'])
Deliverable5_1.head()
```

```python
Deliverable5_1["Profit"] = (Deliverable5_1["SalesPrice"] - Deliverable5_1["PurchasePrice"])
Deliverable5_1 = Deliverable5_1.fillna(0)

Deliverable5_1.head()
```

```python
Deliverable5_2 = Deliverable5_1[["Store", "Profit"]].groupby("Store").sum()
Deliverable5_2.sort_values(by='Profit', ascending=False)
```

```python
Deliverable5_3 = Deliverable5_1[["Brand", "Profit"]].groupby("Brand").sum()
Deliverable5_3.sort_values(by='Profit', ascending=False)
```

This deliverable shows the profit generated for each store and brand for Bibitor. Stores such as 76 and 73 are the best performers, and brands like 1233 and 3535 do the best. We can see this pattern reflected a little bit when we look at beginning and ending inventory with brands such as 1233 and 3545 having high inventories, showing that stores know that these brands are high in demand. The profit numbers give a little more insight on why those inventories might be so high and how it's not necessarily a bad thing to have such high inventory for those products. Overall, it gives a general sense of which stores are doing well and which brands do the best.
