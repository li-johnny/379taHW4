## deliverable 4a

```python
purchases = pd.read_csv("InvoicePurchases12-31-16Sample.csv")
purchases.head(2)
```

```python
purchases["InvoiceDate"] = pd.to_datetime(purchases["InvoiceDate"])
purchases["Year"] = purchases["InvoiceDate"].dt.year
purchases = purchases[(purchases['InvoiceDate'] >= "2016-01-01") & (purchases['InvoiceDate'] <= "2016-12-31")] #filter for 2016
len(purchases)
```

```python
top_vendors = purchases[['VendorNumber','VendorName','Dollars']].groupby(["VendorNumber","VendorName"]).sum()
top_vendors.sort_values(by='Dollars',ascending=False).head(10)
```

## deliverable 4b

```python
invoices = pd.read_csv("InvoicePurchases12-31-16Sample.csv")
invoices = invoices[(invoices['InvoiceDate'] >= "2016-01-01") & (invoices['InvoiceDate'] <= "2016-12-31")]
invoices.head(2)
```

```python
top_freight = invoices[['VendorNumber','VendorName','Freight']].groupby('VendorName').sum()
top_freight.sort_values(by='Freight',ascending=False).head(10)
```

## deliverable 4c

```python
freight_per_dollar = pd.merge(left=top_vendors,
                      right=top_freight,
                      how = 'outer',
                      on = ['VendorName'])
freight_per_dollar.head()
```

```python
freight_per_dollar['FreightPerDollar'] = freight_per_dollar['Freight'] / freight_per_dollar['Dollars']
freight_per_dollar.head(2)
```

```python
freight_per_dollar = freight_per_dollar[freight_per_dollar['Dollars'] >= 250000]
freight_per_dollar.head()
```

```python
freight_per_dollar.sort_values(by='FreightPerDollar',ascending=False).head(10)
```

```python
freight_per_dollar.sort_values(by='FreightPerDollar',ascending=True).head(5)
```

## deliverable 4d

```python
freight_sort = invoices[['VendorNumber','VendorName','Freight','Dollars','Quantity']]
freight_sort.head()
freight_sort = freight_sort[(freight_sort['Quantity'] <= 1000) & (freight_sort['Freight'] > 100)]
freight_sort.groupby('VendorName').sum().sort_values(by='Freight',ascending=False).head()
```
