```python
begInv = pd.read_csv("BegInvFINAL12312016.csv")
```

# Deliverable 5

```python
Sales = pd.read_csv('salesDecember2016Sample.csv')
```

```python
Purchases = pd.read_csv('PurchasesFINAL12312016.csv')
```

```python
Products = Sales[['InventoryId','SalesQuantity','SalesPrice', 'SalesDollars','Store']].groupby('InventoryId').sum()
Products.head()
```

```python
PurProducts = Purchases [['InventoryId','Quantity','Dollars','Store']]
PurProducts.head()
```

```python
PopProducts = pd.merge(left=Products, right=PurProducts, how="outer", on=["InventoryId"]).groupby('InventoryId').sum()
PopProducts.head()
```

```python
PopProducts = PopProducts[['SalesDollars']]
PopProducts.head()
```

```python
PopProducts.sort_values(by='SalesDollars', ascending=False).head(10)
```

Most popular products by amount of sales dollars.
