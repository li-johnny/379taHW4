### Deliverable 4c

```python
invoices['Inventory'] = invoices['Quantity'] * invoices['Dollars']
```

```python
invoices.head()
```

```python
len(invoices)
```

```python
df2016Invoices = invoices[invoices['InvoiceDate']<='2016_12_31'] 
df2016Invoices.head()
```

```python
len(df2016Invoices)
```

```python
dfLargeVendors = df2016Invoices.groupby(['VendorName', 'VendorNumber'])[['Freight','Dollars','Quantity']].sum()
```

```python
dfLargeVendors = dfLargeVendors[dfLargeVendors['Dollars']>=250000]
```

```python
dfLargeVendors.head()
```

```python
dfLargeVendors['FreightRate(%)'] = dfLargeVendors['Freight'] / dfLargeVendors['Dollars'] * 100
dfLargeVendors['FreightPerUnit'] = (dfLargeVendors['Freight'] / dfLargeVendors['Quantity'])
dfLargeVendors.head()
```

#### Lowest 5 Transactions

```python
dfSmallExpensiveInvoices = dfLargeVendors.sort_values(by='FreightRate(%)', ascending = True)
dfSmallExpensiveInvoices.head()
```

#### Top 10 Transactions

```python
dfLargeVendors.sort_values(by='FreightRate(%)', ascending = False).head(10)
```

### Deliverable 4d

```python
dfSmallExp = pd.read_csv("InvoicePurchases12-31-16Sample.csv")
dfSmallExp.head()
```

```python
dfSmallExp = dfSmallExp[dfSmallExp['Quantity']<=1000]
dfSmallExp = dfSmallExp[dfSmallExp['Freight']>=100]
dfSmallExp = dfSmallExp[['VendorNumber','VendorName','Freight','Dollars','Quantity']].groupby(['VendorNumber', 'VendorName']).sum().sort_values(by='Freight', ascending=False)
dfSmallExp.head()
```

### Deliverable 4e

```python
pd.DataFrame([['Freight % of $]',(dfSmallExp['Freight']/dfSmallExp['Dollars']*100).mean(), dfLargeVendors['FreightRate(%)'].mean(),
               (df2016Invoices['Freight']/df2016Invoices['Dollars']*100).mean()], ['Freight per unit', (dfSmallExp['Freight']/dfSmallExp['Quantity']).mean(),
                        dfLargeVendors['FreightPerUnit'].mean(), (df2016Invoices['Freight']/df2016Invoices['Quantity']).mean()]], columns=['', 'Small/Expensive', 'Large Vendors','All Purchases']).set_index('')
```
