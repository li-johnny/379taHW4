## Deliverable 5

```python
dfSalesPrice=pd.read_csv('SalesFINAL12-31-16Sample.csv').groupby(['Description'])[['SalesPrice']].mean()
dfPurchasePrice = purchases.groupby(['Description'])[['PurchasePrice']].mean()
```

```python
dfCM=pd.merge(left=dfSalesPrice, right= dfPurchasePrice, how='inner', on='Description')
dfCM['Contribution Margin']=dfCM['SalesPrice']=dfCM['PurchasePrice']
```

```python
dfCM=dfCM[dfCM['Contribution Margin']<=0]
dfCM.sort_values(by='Contribution Margin', ascending=True)
```
