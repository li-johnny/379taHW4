```python
vendorsD = vendors[['VendorNumber','VendorName','Quantity','Dollars','Freight']]
vendorsD.head(2)
```

```python
print (vendorsD.dtypes)
```

```python
vendorsD = pd.DataFrame(vendorsD)
```

```python
vendorsD['Quantity']=vendorsD['Quantity'].astype(float)
```

```python
vendorsD = vendorsD[(vendorsD['Quantity']<=1000 ) | (vendorsD['Freight']>100)]
vendorsD.sort_values(by='Freight', ascending=False).head()
```

```python
vendorsD = vendorsD.groupby('VendorName')[['Quantity', 'Dollars','Freight']].sum()
```

```python
vendorsD.sort_values(by="Freight",ascending=False).head(10) #Returns highest freight values, given the restrictions
```
