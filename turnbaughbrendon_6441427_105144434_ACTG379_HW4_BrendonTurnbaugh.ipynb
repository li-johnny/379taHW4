# Deliverable 4a

```python
purchases = pd.read_csv("InvoicePurchases12-31-16Sample.csv")
```

```python
purchases = purchases.rename(columns = {'Dollars':'InvCost'})
purchases['TotalCost'] = purchases['InvCost'] + purchases['Freight']
```

```python
purchases.head(2)
```

```python
purchasesv = purchases[['VendorNumber','VendorName','TotalCost']].groupby('VendorName').sum()
```

```python
purchasesv.sort_values(by='TotalCost',ascending=False).head(10)
```

# Deliverable 4b

```python
purchases4b = purchases[['VendorName','Freight']].groupby('VendorName').sum()
```

```python
purchases4b.sort_values(by='Freight',ascending=False).head(10)
```

# Deliverable 4c

```python
purchases4c = pd.merge(purchasesv,purchases4b,how = 'outer', on = 'VendorName')
```

```python
purchases4c['FreightPerDollar'] = purchases4c['Freight']/purchases4c['TotalCost']
```

```python
purchases4c.head()
```

```python
purchases4cc = purchases4c[purchases4c['TotalCost'] > 250000]
```

```python
purchases4cc.sort_values(by='FreightPerDollar',ascending=False).head(10)
```

```python
purchases4cc.sort_values(by='FreightPerDollar',ascending=False).tail(5)
```

# Deliverable 4d

```python
purchases4d = purchases[['Quantity','Freight','VendorName','TotalCost']]
```

```python
purchases4d.head()
```

```python
purchases4d = purchases4d[(purchases4d['Quantity'] <= 1000) & (purchases4d['Freight'] > 100)]
```

```python
purchases4d.head()
```

```python
purchases4d.sort_values(by='Freight',ascending=False).head(10)
```

# Deliverable 4e

```python
purchases4d['FreightPerDollar'] = purchases4d['Freight']/purchases4d['TotalCost']
```

```python
purchases4d['FreightPerUnit'] = purchases4d['Freight']/purchases4d['Quantity']
```

```python
purchases4d.sort_values(by='FreightPerDollar',ascending=False).head(10)
```
