## Deliverable #4b

## Deliverable #4c

Calculate freight per dollar purchased for only those Vendors who Bibitor purchased at least 250,000 of inventory during 2016 and identify the ten Vendors with the highest per dollar freight and the five Vendors with the lowest per $ freight.

```python
freight = invoices.groupby(["VendorName","VendorNumber"])[["Freight", "Dollars"]].sum()
freight = freight[freight["Dollars"]>=250000]
freight["Freight per Dollar"] = (freight["Freight"]/freight["Dollars"])
```

```python
freight = freight.sort_values(by='Freight per Dollar', ascending=False)
freight.head(10)
```

```python
freight = freight.sort_values(by='Freight per Dollar', ascending=False)
freight.tail(5)
```

## Deliverable #4d

Identify those transactions where the freight cost was greater than $100 and the 
units were less than or equal to 1,000 units. Which Vendors had the highest Freigh 
cost under these conditions.

```python
freight = freight[freight["Freight"]>100]
freight = invoices[invoices["Quantity"]<=1000]
freight = freight.groupby(["VendorName", "VendorNumber"])[["Freight","Dollars","Quantity"]].sum().sort_values(by="Freight", ascending=False)
freight.head()
```

## Deliverable #4e

Calculate the freight per dollar & per unit for those transactions in d. How do they 
compare to your findings in c above?

```python
freight["Freight per Dollar"] = (freight['Freight']/freight['Dollars'])
freight["Freight per Unit"] = (freight['Freight']/freight['Quantity'])
freight = freight.groupby(["VendorName", "VendorNumber"])[["Freight","Dollars","Freight per Dollar","Freight per Unit"]].sum().sort_values(by="Freight", ascending=False)
freight.head()
```

The companies are selling different amounts of product but are experiencing similar freight costs.
