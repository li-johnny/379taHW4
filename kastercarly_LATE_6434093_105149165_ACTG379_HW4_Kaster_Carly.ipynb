# Deliverable 1

```python
BegInv['InvCost_Beg'] = BegInv['onHand_Beg'] * BegInv['Price']
BegInv.head()
```

```python
BegInv = BegInv.rename(columns = {'onHand':'onHand_Beg'})
BegInv.head(2)
```

```python
BegInv_Store = BegInv[['Store','onHand_Beg','InvCost_Beg']].groupby('Store').sum()
BegInv_Store.head()
```

```python
BegInv_Store.sort_values(by='InvCost_Beg',ascending=False).head(10)
```

```python
BegInv_Brand = BegInv[['Brand','onHand_Beg','InvCost_Beg']].groupby('Brand').sum()
BegInv_Brand.head()
```

```python
BegInv_Brand.sort_values(by='InvCost_Beg',ascending=False).head(10)
```

# Deliverable 2

```python
EndInv['InvCost_End'] = EndInv['onHand_End'] * EndInv['Price']
EndInv.head()
```

```python
EndInv_Store.sort_values(by='InvCost_End',ascending=False).head(10)
```

```python
EndInv_Brand.sort_values(by='InvCost_End',ascending=False).head(10)
```
