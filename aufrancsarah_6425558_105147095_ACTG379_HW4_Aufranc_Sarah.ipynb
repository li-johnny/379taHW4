# Deliverable 5

```python
sales = pd.read_csv('salesDecember2016Sample.csv')
sales.head(1)
```

```python
invoices = pd.read_csv('2017PurchasePricesDecSample.csv')
invoices.head(1)
```

```python
Sales = sales[['Description','SalesQuantity','SalesDollars']]
Sales = Sales.groupby(['Description']).sum()
Sales.head(1)
```

```python
Invoices = invoices[['Description','PurchasePrice']]
Invoices = Invoices.groupby(['Description']).sum()
Invoices.head(1)
```

```python
SalesInvoices = pd.merge(left=Sales, right=Invoices, how='outer', on=['Description'])
SalesInvoices.head(1)
```

```python
SalesInvoices['AverageSalesPrice'] = (SalesInvoices['SalesDollars'] / SalesInvoices['SalesQuantity'])
SalesInvoices.head(1)
```

```python
SalesInvoices['PurchasesDollars'] = (SalesInvoices['PurchasePrice'] * SalesInvoices['SalesQuantity'])
SalesInvoices.head(1)
```

```python
SalesInvoices['CMperUnit'] = (SalesInvoices['AverageSalesPrice'] - SalesInvoices['PurchasePrice'])
SalesInvoices.head(1)
```

```python
SalesInvoices['TotalCM'] = (SalesInvoices['SalesDollars'] - SalesInvoices['PurchasesDollars'])
SalesInvoices.head(1)
```

```python
SalesInvoicesTotal = SalesInvoices.sort_values(by='TotalCM', ascending=False)
SalesInvoicesTotal.head()
```

```python
SalesInvoicesUnit = SalesInvoices.sort_values(by='CMperUnit', ascending=False)
SalesInvoicesUnit.head()
```
