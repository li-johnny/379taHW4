# 4.C

```python
FreightMath = Invoice_Purchases[['VendorName','Freight','Dollars','Quantity',]].groupby('VendorName').sum()
```

```python
FreightMath['Freight/Dollar%'] = (FreightMath['Freight'] / FreightMath['Dollars'] *100).round(2)
```

```python
FreightMath['Freight/Unit'] = (FreightMath['Freight'] / FreightMath['Quantity']).round(2)
```

```python
FreightMathCriteria = FreightMath[FreightMath["Dollars"] > 250000]
```

```python
FreightMathCriteria.sort_values(by='Freight/Dollar%', ascending=False).head(10)
```

```python
FreightMathCriteria.sort_values(by='Freight/Dollar%', ascending=True).head()
```

# 4.D

```python
deliverable4Dfilter = Invoice_Purchases[Invoice_Purchases["Freight"] > 100]
deliverable4Dfilter = deliverable4Dfilter[deliverable4Dfilter["Quantity"] <= 1000]
```

```python
deliverable4Dfilter.head()
```

```python
deliverable4D = deliverable4Dfilter[['VendorNumber','Freight','Dollars','Quantity',]].groupby('VendorNumber').sum()
deliverable4D.head(1)
```

# 4.E

```python
FreightMath['Freight/Dollar%'] = (FreightMathCriteria["Freight"]/FreightMathCriteria["Dollars"])
FreightMath['Freight/Unit'] = (FreightMath["Freight"]/FreightMath["Quantity"]*100)
Invoice_Purchases['Freight/Unit'] = (Invoice_Purchases["Freight"]/Invoice_Purchases["Quantity"])
Invoice_Purchases['Freight/Dollar%'] = (Invoice_Purchases["Freight"]/Invoice_Purchases["Dollars"]*100)
```

```python
pd.DataFrame([['Freight/Dollar%', FreightMath["Freight/Dollar%"].mean(), FreightMath["Freight/Unit"].mean(),
               Invoice_Purchases["Freight/Dollar%"].mean()],["Freight/Unit", FreightMath["Freight/Unit"].mean(),
                FreightMathCriteria['Freight/Dollar%'].mean(), Invoice_Purchases["Freight/Unit"].mean()]],
             columns = [" ", "Small", "Large", "Average"]).set_index(" ").round(2)
```
