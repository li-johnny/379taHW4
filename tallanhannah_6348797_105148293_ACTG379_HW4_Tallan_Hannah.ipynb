## Deliverable 4e
_______________________________________________________________________________________________________

```python
SmallExpensive = fourInvoices[(fourInvoices["Freight"] > 100) & (fourInvoices["Quantity"] <= 1000)]

SmallExpensive['Per$Freight'] = SmallExpensive['Freight'] / SmallExpensive['Dollars']

SmallExpensive['PerUnitFreight'] = SmallExpensive['Freight'] / SmallExpensive['Quantity']

SmallExpensive.head()
```

```python
LargeVend = fourInvoices.groupby(["VendorNumber","VendorName"])[["Freight","Dollars","Quantity"]].sum()

LargeVend = LargeVend[LargeVend["Dollars"]>250000]

LargeVend['Per$Freight'] = LargeVend['Freight'] / LargeVend['Dollars']

LargeVend['PerUnitFreight'] = LargeVend['Freight'] / LargeVend['Quantity']

LargeVend = LargeVend[LargeVend["Dollars"]>250000]

LargeVend.head()
```

```python
avg_small_expensive = SmallExpensive[['Per$Freight', 'PerUnitFreight']].mean()
avg_large_vend = LargeVend[['Per$Freight', 'PerUnitFreight']].mean()

MeanTable = pd.DataFrame([avg_small_expensive, avg_large_vend], index=['Small and Expensive', 'Large Vendors'])

MeanTableTransp = MeanTable.T

MeanTableTransp
```

Comparing these two, I see that the small/expensive vendors have a much higher freight rate (for quantity and dollars) relative to the large vendors, who have a relatively smaller freight rate (for quantity and dollars). This makes sense, as larger orders might get better freight prices/bulk discount shipping deals. 

## Deliverable 5: Freeform Sales and Purchases data combination and analysis. 
_______________________________________________________________________________________________________

My goal is to show the difference by brand in amounts recorded as received in Sales data versus dollars invoiced from the Invoice Purchases data. I am making the simplified assumption here that cash inflow from sales would be the former, and the difference between the two would be accounts receivable. This might be helpful to show the brands most purchased on credit by dollar amount.

```python
Sales = pd.read_csv("SalesFINAL12312016.csv")

ByVenSales = Sales.groupby(["VendorNo","VendorName"])[["SalesDollars"]].sum().sort_values(by="SalesDollars", ascending=False)

ByVenSales.rename(columns={'SalesDollars': 'SalesDollarsRecorded'}, inplace=True)

ByVenSales.head()
```

```python
InvPurch = pd.read_csv("InvoicePurchases12312016.csv")

ByVenInvPurch = InvPurch.groupby(["VendorNumber","VendorName"])[["Dollars"]].sum().sort_values(by="Dollars", ascending=False)

ByVenInvPurch.rename(columns={'Dollars': 'DollarsInvoiced'}, inplace=True)

ByVenInvPurch.head()
```

```python
mergedVens = pd.merge(left=ByVenInvPurch, right=ByVenSales, how='outer', on=['VendorName'])

mergedVens['InvoicedOverSale'] = mergedVens['DollarsInvoiced'] - mergedVens['SalesDollarsRecorded']

mergedVens.sort_values(by="InvoicedOverSale", ascending=False).head()


```

The above 5 brands represent the top 5 that A/R balances can be attributed to.
