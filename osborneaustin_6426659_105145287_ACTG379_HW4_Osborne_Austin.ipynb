```python
sales = pd.read_csv("salesDecember2016Sample.csv").groupby(["Description", "VendorName"], as_index=False)[["SalesPrice"]].sum()
purchases = pd.read_csv("2017PurchasePricesDecSample.csv").groupby(["Description", "VendorName"], as_index=False)[["PurchasePrice"]].sum()
top_Three = purchases[["VendorName", "Description", "PurchasePrice"]].groupby("VendorName").sum()
top_Three_Main = top_Three.sort_values(by="PurchasePrice",ascending=False).head(3)
top_Three_Main
```

```python
sp_Merge = pd.merge(left = sales, right = purchases, how = "inner", on = "Description")
sp_Merge["Contribution Margin"] = sp_Merge["SalesPrice"] - sp_Merge["PurchasePrice"]
main_Merge = sp_Merge[["Contribution Margin", "VendorName_x"]]
main_Merge.head(10)
```

```python
main_Merge = main_Merge.groupby(["VendorName_x"],as_index=False)[["Contribution Margin"]].sum()
CM = main_Merge[(main_Merge["VendorName_x"] == "ULTRA BEVERAGE COMPANY LLP") | (main_Merge["VendorName_x"] == "PERFECTA WINES") | (main_Merge["VendorName_x"] == "MARTIGNETTI COMPANIES")]
CM_Main = CM.sort_values(by="Contribution Margin", ascending=False)
CM_Main
```

This code first identifies the top three suppliers that Bibitor received inventory from (in dollars) and the total purchase price for this inventory in December.

Then the code merges the December purchase and sales data and generates a contribution margin that the company received from selling inventory from all suppliers.

Finally, the code groups the vendors and the sum of each vendors contribution margin, selects only the top three suppliers that Bibitor received inventory from (in dollars), and projects the contribution margin received from the sales from these suppliers.
