```python
MergedStoreEndInv = MergedStoreEndInv.drop_duplicates(subset=['Store'])
```

```python
MergedStoreEndInv["Total Value"] = (MergedStoreEndInv["SalesPrice"]) + (MergedStoreEndInv["EndInv"] * MergedStoreEndInv["EndInvCost"])
```

```python
MergedStoreEndInv.sort_values(by='Total Value').tail(5).sort_values(by='Total Value', ascending = False)
```

This gives us information on how many sales + end value of inventory we have on hand. This can be useful to managers to view the amount of money each store is making before calculating the expenses from each. This way they can see where they might in the future need to attempt to cut expenses.
