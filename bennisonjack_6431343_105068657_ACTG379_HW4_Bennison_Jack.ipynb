# Deliverable 1a

```python
begInv = pd.read_csv("BegInvFINAL12-31-16Sample.csv")
```

```python
begInv.head()
```

```python
begInv = begInv.rename(columns = {'onHand' : 'BegQuantity'})
begInv.head()
```

```python
begInv['BegInvCost'] = begInv['BegQuantity'] * begInv['Price']
begInv.head()
```

```python
begInv_Store = begInv[['Store','BegQuantity','BegInvCost']].groupby('Store').sum()
begInv_Store.head()
```

```python
begInv_Store.sort_values(by='BegInvCost',ascending=False).head(10)
```

# Deliverable 4a

```python
Purchases = pd.read_csv("PurchasesFINAL12-31-16Sample.csv")
Purchases.head()
```

```python
Topvendors = Purchases[['VendorNumber','VendorName','Dollars']].groupby('VendorNumber').sum()
```

```python
Topvendors.sort_values(by='Dollars',ascending=False).head(10)
```

# Deliverable 4b

```python
InvoicePurchases = pd.read_csv("InvoicePurchases12-31-16Sample.csv")
InvoicePurchases.head()
```

```python
Topvendorsfreight = InvoicePurchases[['VendorNumber','VendorName','Freight']].groupby('VendorNumber').sum()
Topvendorsfreight.sort_values(by='Freight',ascending=False).head(10)
```

# Deliverable 4c

```python
Largepurchases = Topvendors[Topvendors['Dollars'] > 250000]
Largepurchases.head()
```

```python
PurchasesandInvoices = pd.merge(left = Largepurchases, right = Topvendorsfreight[['Freight']], how = 'inner', on = ['VendorNumber'])
```

```python
PurchasesandInvoices['Freightper$'] = PurchasesandInvoices['Freight'] / PurchasesandInvoices['Dollars']
```

```python
PurchasesandInvoices.head()
```

```python
PurchasesandInvoices.sort_values(by='Freightper$',ascending=False).head(10)
```

```python
PurchasesandInvoices.sort_values(by='Freightper$',ascending=True).head(5)
```

# Deliverable 4d

```python
Freightcostunits = InvoicePurchases[(InvoicePurchases['Quantity'] <= 1000) & (InvoicePurchases['Freight'] > 100)]
Freightcostunits.head()
```

```python
Freightcostunits.sort_values(by='Freight',ascending=False).head()
```

```python
#These are the vendors with the highest freight costs under these conditions
```

# Deliverable 4e

```python
Freightcostunits.loc[:, 'Freightper$'] = Freightcostunits['Freight'] / Freightcostunits['Dollars']
Freightcostunits.loc[:, 'Freightperunit'] = Freightcostunits['Freight'] / Freightcostunits['Quantity']
Freightcostunits.head()
```

```python
Freightcostunits.sort_values(by='Freightper$',ascending=False).head()
```

```python
#The freight per $ is much higher here than in part c
```
