```python
dfSalesData=pd.read_csv("salesDecember2016Sample.csv")
```

```python
dfPurchasePrice = dfPurchases.groupby(['Description'])[['PurchasePrice']].mean()
```

```python
dfContributionMargin = pd.merge(left = dfSalesData, right = dfPurchasePrice, how='inner', on='Description')
```

```python
dfContributionMargin['Contribution Margin'] = dfContributionMargin['SalesPrice'] - dfContributionMargin['PurchasePrice']
```

```python
dfContributionMargin = dfContributionMargin[dfContributionMargin["Contribution Margin"] <=0]
```

```python
dfContributionMargin.sort_values(by='Contribution Margin', ascending=True)
```
