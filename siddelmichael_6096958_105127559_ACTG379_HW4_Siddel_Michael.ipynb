## Deliverable 5
#### I will write code that combines the sales and purchases data to return the total contribution margin earned on each product sold in March 2016. For each product, contribution margin will be calculted by taking total sales revenue and subtracting total the total purchase prices for transactions in March 2016. I will be ignoring freight costs and excise taxes.

```python
pur16 = purchases[(purchases["InvoiceDate"] >= "2016-03-01") & (purchases["InvoiceDate"] <= "2016-03-31")]
sal16 = sales[(sales["SalesDate"] >= "2016-03-01") & (sales["SalesDate"] <= "2016-03-31")]
cm = pd.merge(left = purchases, right = sal16, how = "inner", on = "InventoryId")
cm["Purchase Dollars"] = (cm["PurchasePrice"] * cm["SalesQuantity"])
cm["Contribution Margin"] = (cm["SalesDollars"] - cm["Purchase Dollars"])
cm = cm.groupby(["InventoryId"])[["Contribution Margin"]].sum()
cm = cm.sort_values(by = "Contribution Margin", ascending = False)
cm.head(10)
```
