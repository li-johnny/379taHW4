## Deliverable 4C

```python
invoices['Inventory']=invoices['Quantity']*invoices['Dollars']
```

```python
invoices.head()
```

```python
df2016Invoices = invoices[(invoices['InvoiceDate']>='2016-01-01')&(invoices['InvoiceDate']<='2016-12-31')]
```

```python
df2016Invoices.head()
```

```python
dfLargeVendors = df2016Invoices.groupby(['VendorNumber', 'VendorName'])[['Freight', 'Dollars', 'Quantity']].sum()
```

```python
dfLargeVendors = dfLargeVendors[dfLargeVendors['Dollars']>=250000]
```

```python
dfLargeVendors.head()
```

```python
dfLargeVendors['FreightRate(%)']=(dfLargeVendors['Freight']/dfLargeVendors['Dollars'])*100
dfLargeVendors['FreightPreUnit']=(dfLargeVendors['Freight']/dfLargeVendors['Quantity'])
```

```python
dfLargeVendors.head()
```

```python
dfLargeVendors.sort_values(by='FreightRate(%)', ascending=True).head(5)
```

```python
dfLargeVendors.sort_values(by='FreightRate(%)', ascending=False).head(10)
```
