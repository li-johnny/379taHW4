# Deliverable 3

```python
tbl5 = pd.merge(left = beginv, right = endinv, how = "outer", on = "InventoryId")
tbl5 = tbl5.groupby(["InventoryId"])[["onHand_x", "InvCost01_01_2016", "onHand_y", "InvCost12_31_2016"]].sum()
tbl5 = tbl5.rename(columns = {"onHand_x": "Beginning Quantity"})
tbl5 = tbl5.rename(columns = {"onHand_y": "Ending Quantity"})
tbl5 = tbl5.rename(columns = {"InvCost01_01_2016": "Beginning Total"})
tbl5 = tbl5.rename(columns = {"InvCost12_31_2016": "Ending Total"})
tbl5.head()
```
