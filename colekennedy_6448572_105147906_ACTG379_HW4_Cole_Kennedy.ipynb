# Deliverable 4

## Part A

```python
Purchases = pd.read_csv('PurchasesFINAL12-31-16Sample.csv')
Purchases.head(2)
```

```python
PurchasesA = Purchases[(Purchases['InvoiceDate']>='2016-01-01') & (Purchases['InvoiceDate']<='2016-12-31')]
PurchasesA.head(2)
```

```python
VendorsA = PurchasesA[['VendorNumber', 'VendorName', 'Dollars']].groupby(['VendorNumber', 'VendorName']).sum()
VendorsA.head(2)
```

```python
VendorsA.sort_values(by='Dollars', ascending = False).head(10)
```

## Part B

```python
Purchases2 = pd.read_csv('InvoicePurchases12-31-16Sample.csv')
Purchases2.head(2)
```

```python
PurchasesB = Purchases2[(Purchases2['InvoiceDate']>='2016-01-01') & (Purchases2['InvoiceDate']<='2016-12-31')]
PurchasesB.head(2)
```

```python
FreightB = PurchasesB[['VendorNumber', 'VendorName', 'Freight']].groupby(['VendorNumber', 'VendorName']).sum()
FreightB.head(2)
```

```python
FreightB.sort_values(by='Freight', ascending = False).head(10)
```

## Part C

```python
VendorsC = PurchasesB[['VendorNumber', 'VendorName', 'Dollars', 'Freight']].groupby(['VendorNumber', 'VendorName']).sum()
VendorsC.head(2)
```

```python
Vendor_over_250 = VendorsC[VendorsC['Dollars'] > 249999]
Vendor_over_250.head(2)
```

```python
Vendor_over_250['FreightPerDollarPercent'] = (Vendor_over_250['Freight'] / Vendor_over_250['Dollars']) * 100
Vendor_over_250.head(2)
```

```python
Vendor_over_250.sort_values(by='FreightPerDollarPercent', ascending = False).head(10)
```

```python
Vendor_over_250.sort_values(by='FreightPerDollarPercent', ascending = False).tail(5)
```

## Part D

```python
TransactionsD = PurchasesB[(PurchasesB['Quantity']<= 1000) & (PurchasesB['Freight']>= 100)]
TransactionsD.head(2)
```

```python
VendorsD = TransactionsD[['VendorNumber', 'VendorName', 'Quantity', 'Freight', 'Dollars']].groupby(['VendorNumber', 'VendorName']).sum()
VendorsD.head(2)
```

```python
VendorsD.sort_values(by='Freight', ascending = False).head(10)
```

## Part E

```python
VendorsD['FreightPerDollarPercent'] = (VendorsD['Freight'] / Vendor_over_250['Dollars']) * 100
VendorsD.head(2)
```

```python
VendorsD['FreightPerUnit'] = VendorsD['Freight'] / VendorsD['Quantity']
VendorsD.head(2)
```

```python
VendorsD.sort_values(by='Freight', ascending = False).head(10)
```

The freight per dollar percentage for the transactions in Part D is significantly lower than the freight per dollar percentage of the transactions in Part C. Even the highest freight per dollar percentage (0.16%) of the transactions in Part D is lower than the lowest freight per dollar percentage of the transactions in Part C. This indicates that it costs the company more comparatively to ship with smaller shippers/loads than with larger shippers/loads.
