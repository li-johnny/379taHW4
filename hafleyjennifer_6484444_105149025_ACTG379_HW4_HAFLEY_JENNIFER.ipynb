# Deliverable 4b.

```python
Inv_Purch = pd.read_csv('OneDrive - Oregon State University/ACTG379/ACTG Analytics/HW4 Sampled Data/InvoicePurchases12-31-16Sample.csv')
Inv_Purch.head()
```

```python
Vendors_Freight = Inv_Purch[["VendorNumber","VendorName","Freight"]].groupby(["VendorNumber","VendorName"]).sum().sort_values(by="Freight", ascending = False)
Vendors_Freight.head(10)
```

# Deliverable 4c.

```python
Large_Purch = Inv_Purch[Inv_Purch['Dollars']>=250000]
Large_Purch.head()
```

```python
Large_Purch_2 = Large_Purch[["VendorNumber", "VendorName", "Dollars", "Freight"]].groupby(["VendorNumber", "VendorName"]).agg({"Freight":['sum'], "Dollars":['sum']})
Large_Purch_2['Freight Ratio'] = (Large_Purch_2["Freight"].div(Large_Purch_2["Dollars"]))
Large_Purch_2 = Large_Purch_2.fillna(0)
Large_Purch_2.nlargest(10, 'Freight Ratio')
```

```python
Large_Purch_2.nsmallest(5,'Freight Ratio')
```

# Deliverable 4d.

```python
FC_Unit= Inv_Purch[(Inv_Purch['Freight']>100) & (Inv_Purch['Quantity']<=1000)]
FC_Unit.head()
```

```python
FC_Unit_2 = FC_Unit[["VendorNumber","VendorName","Freight","Quantity","Dollars"]].groupby(["VendorNumber","VendorName"]).sum().sort_values(by="Freight", ascending=False)
FC_Unit_2.head()
```

# Deliverable 4e.

```python
FC_Unit_2['Freight Per Dollar'] = (FC_Unit_2["Freight"].div(FC_Unit_2["Dollars"]))
FC_Unit_2['Freight Per Unit'] = (FC_Unit_2['Freight'].div(FC_Unit_2['Quantity']))
FC_Unit_2= FC_Unit_2.fillna(0)
FC_Unit_2.head()
```

# Deliverable 5

```python
Sales = pd.read_csv("OneDrive - Oregon State University/ACTG379/ACTG Analytics/HW4 Sampled Data/InvoicePurchases12-31-16Sample.csv")
Sales.head(100)
```

```python
Sales['date'] = pd.to_datetime(Sales['InvoiceDate'])
Sales['Month'] = Sales['date'].dt.month_name()
Sales.head()
```

```python
order = "January February March April May June July August September October November December".split() 

month_df = Sales[["Quantity", "Month"]].groupby(['Month']).mean().sort_values(by="Quantity")
month_df = month_df.loc[order]
month_df.head(12)
```
