# Deliverable 5

```python
salesdata = pd.read_csv("SalesFINAL12-31-16Sample.csv").groupby(["InventoryId"],as_index=False)[["SalesPrice"]].mean()
```

```python
salesdata.head()
```

```python
purchases = pd.read_csv("PurchasesFINAL12-31-16Sample.csv").groupby(["InventoryId"],as_index=False)[["PurchasePrice"]].mean()
```

```python
purchases.head()
```

```python
mergeA = pd.merge(left=salesdata, right=purchases, how='outer', on=["InventoryId"])
mergeA.head()
```

```python
mergeA["ContributionMargin"] = (mergeA["SalesPrice"] - mergeA["PurchasePrice"])
```

```python
mergeA.head()
```

```python
mergeA = mergeA[['SalesPrice','PurchasePrice','ContributionMargin','InventoryId']].groupby('InventoryId').sum()
```

```python
mergeA.head()
```

```python
mergeA =mergeA[mergeA["ContributionMargin"]<=0]
mergeA.sort_values(by="ContributionMargin", ascending = True)
```

I have choosen to find the contribution margin from the sales data set to the purchases data set. I then sort them by the highest negative amounts of contribution margin based on products (InventoryId)
