## Deliverable 4C

```python
P25F= vendorS[vendorS['Dollars'] >= 250000]
P25= P25F[['VendorNumber','VendorName','Freight','Dollars']].groupby(['VendorNumber','VendorName']).sum()
P25.head(5)
```

```python
P25['Freightper$'] = (P25['Freight'] / P25['Dollars'])
```

```python
#Highest
P25_sorted = P25.sort_values(by='Freightper$', ascending=False)
P25_sorted.head(10)
```

```python
#Lowest
P25_sorted2 = P25.sort_values(by='Freightper$', ascending=True)
P25_sorted2.head(5)
```

## Deliverable 4D

```python
#FT = Filtered Transactions
FT = vendorS[(vendorS['Freight'] >= 100) & (vendorS['Quantity'] <= 1000)]
FreightSum= FT[['VendorNumber', 'VendorName','Freight','Quantity']].groupby(['VendorNumber','VendorName']).sum()
FS_sorted= FreightSum.sort_values(by='Freight', ascending=False)
FS_sorted.head(10)
```

## Deliverable 4E

```python
FreightSum= FT[['VendorNumber', 'VendorName','Freight','Quantity','Dollars']].groupby(['VendorNumber','VendorName']).sum()
FreightSum['Freightper$'] = (FreightSum['Freight'] / FreightSum['Dollars'])
FS_sorted= FreightSum.sort_values(by='Freightper$', ascending=False)
FS_sorted.head(10)
```

The Freight per Dollars when filtered for the vendors who had more than 100 freight and less than 1000 quanitity seems to be marketably lower than the vendors filtered strcitly based on more than $250,000 in purchases. The freight per dollar in part d is almost all lower than the lowest values in part C.
