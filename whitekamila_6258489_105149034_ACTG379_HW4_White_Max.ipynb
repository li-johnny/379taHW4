### Deliverable 3

Create a table that shows both the beginning and ending inventory balances in terms of quantity (# of units) and total $$ value (total cost of inventory) held for each Inventory Id. This table should have five columns: InventoryId, Beginning Quantity, Beginning Total Dollar (ie., quantity * price), Ending Quantity, and Enging Total $ (i.e., quantity * price)

```python
begInvMerge = begInv[['InventoryId','01_01_2016_Qty','InvCost01_01_2016']].groupby('InventoryId').sum()
begInvMerge['Beginning Quantity'] = begInvMerge['01_01_2016_Qty']
begInvMerge['Beginning Total $'] = begInvMerge['InvCost01_01_2016']
begInvMerge = begInvMerge.drop('01_01_2016_Qty',axis = 1)
begInvMerge = begInvMerge.drop('InvCost01_01_2016',axis = 1)

endInvMerge = endInv[['InventoryId','01_01_2016_Qty','InvCost01_01_2016']].groupby('InventoryId').sum()
endInvMerge['Ending Quantity'] = endInvMerge['01_01_2016_Qty']
endInvMerge['Ending Total $'] = endInvMerge['InvCost01_01_2016']
endInvMerge = endInvMerge.drop('01_01_2016_Qty',axis = 1)
endInvMerge = endInvMerge.drop('InvCost01_01_2016',axis = 1)

totInv = pd.merge(left=begInvMerge,right=endInvMerge,how='outer', on = ['InventoryId'])
totInv= totInv.fillna(0)
totInv.head()
```

### Deliverable 4

a. Identify the top ten Vendors (i.e., those 10 Vendors who sold Bibtor the most inventory in total $ during 2016).

```python
dfSales = pd.read_csv("PurchasesFINAL12-31-16Sample.csv")
dfSales = dfSales[['Quantity','VendorName']].groupby('VendorName').sum() #Volume meaning how much they sold
dfSales.sort_values(by='Quantity',ascending=False).head(10)
```

b. Identify the ten Vendors who charged the most freight during 2016.

```python
dfFreight = pd.read_csv("InvoicePurchases12-31-16Sample.csv")
dfFreight2 = dfFreight[['Freight','VendorName']].groupby('VendorName').sum() #Volume meaning how much they sold
dfFreight2.sort_values(by='Freight',ascending=False).head(10)
```

c. Calculate freight per $ purchased for only those Vendors who Bibitor purchased at least 250,000 of inventory during 2016 and identify the ten Vendors with the highest per dollar freight and the five Vendors with the lowest per dollar freight.

```python
dfFreight3 = dfFreight[['Freight','VendorName','Dollars']].groupby('VendorName').sum()
dfFreight3['Per Dollar Freight'] = dfFreight3['Freight']/dfFreight3['Dollars']
dfFreight4 = dfFreight3[(dfFreight3["Dollars"]>=250000)]
print("The ten Vendors with the highest per dollar freight are:",dfFreight4.sort_values(by='Per Dollar Freight',ascending=False).head(10))
print("The five Vendors with the lowest per dollar freight are:",dfFreight4.sort_values(by='Per Dollar Freight',ascending=True).head(5))
```

d. Identify those transactions where the freight cost was greater than $100 and the units were less than or equal to 1,000 units. Which Vendors had the highest Freight cost under these conditions.

```python
dfFreight5 = dfFreight[(dfFreight["Freight"]>100) & (dfFreight["Quantity"]<=1000)] #Cost by transaction
dfFreight6 = dfFreight5[['Freight','VendorName','Quantity']].groupby('VendorName').sum() #Total cost
dfFreight6.sort_values(by='Freight',ascending = False).head(10)
```

e. Calculate the freight per $ & per unit for those transactions in d. How do they compare to your findings in c above?

```python
dfFreight7 = dfFreight5[['Freight','VendorName','Quantity','Dollars']].groupby('VendorName').sum()
dfFreight7['Freight per $'] = dfFreight7['Freight']/dfFreight7['Dollars']
dfFreight7['Freight per unit'] = dfFreight7['Freight']/dfFreight7['Quantity']
dfFreight7.sort_values(by='Freight per $',ascending = False).head(10)

#The per dollar prices are similar but the vendors are different. 
#It makes sense, as the vendors in C might be purchasing more than $100 or 1,000 per transaction.
```

### Deliverable 5

Using the sales data and at least some of the other data, do an analysis of your choice. If you have trouble with the sales data file size, you can use the December only sales data.

```python
dfDec16 = pd.read_csv("salesDecember2016Sample.csv")
```

```python
#By at least some of the other data, do you mean another file?
dfDec17 = pd.read_csv("2017PurchasePricesDecSample.csv")
```

```python
dfDec16['Price 2016'] = dfDec16['SalesPrice']
dfLeft = dfDec16[['Brand','Price 2016']].groupby('Brand').mean()
dfDec17['Price 2017'] = dfDec17['Price']
dfRight = dfDec17[['Brand','Price 2017']].groupby('Brand').mean()
dfDec = pd.merge(left=dfLeft,right=dfRight,how='inner', on = ['Brand'])
dfDec['Change in Price'] = dfDec['Price 2017'] - dfDec['Price 2016']
print("The ten brands with the biggest average price change from 2016 to 2017 are:",dfDec.sort_values(by='Change in Price',ascending=False).head(10))
```
