```python
freight_df = pd.read_csv(r"C:/Users/keith/Downloads/HW4/Data/HW4 Sampled Data/InvoicePurchases12-31-16Sample.csv")
```

```python
freight_df_b = freight_df.groupby(['VendorName', 'VendorNumber']).sum()

freight_df_b = freight_df_b.sort_values(by='Freight', ascending=False)
freight_df_b = freight_df_b[['Freight']]
freight_df_b.head(10)
```

```python
freight_df_c = freight_df.groupby(['VendorName', 'VendorNumber']).sum()


freight_df_c = freight_df_c[['Freight', 'Dollars', 'Quantity']]
freight_df_c['FreightRate(%)'] = (freight_df_c['Freight'] / freight_df_c['Dollars']) * 100
freight_df_c['FreightPerUnit'] = (freight_df_c['Freight'] / freight_df_c['Quantity'])
freight_df_c = freight_df_c.sort_values(by='FreightRate(%)', ascending=False)
freight_df_c['FreightRate(%)'] = freight_df_c['FreightRate(%)'].round(2)
freight_df_c['FreightPerUnit'] = freight_df_c['FreightPerUnit'].round(2)
freight_df_c = freight_df_c[freight_df_c['Dollars'] > 250000]
freight_df_c.head(10)
```

```python
freight_df_d = freight_df.groupby(['VendorName', 'VendorNumber']).sum()
freight_df_d = freight_df_d.sort_values(by='Dollars', ascending=False)
freight_df_d = freight_df_d[freight_df_d['Dollars'] > 100]
freight_df_d = freight_df_d[freight_df_d['Quantity'] <= 1000]
freight_df_d = freight_df_d[['Freight', 'Dollars', 'Quantity']]

freight_df_d.head(5)
```

```python
freight_df_e = freight_df.groupby(['VendorName', 'VendorNumber']).sum()
freight_df_e['FreightRate(%)'] = (freight_df_e['Freight'] / freight_df_e['Dollars']) * 100
freight_df_e['FreightPerUnit'] = (freight_df_e['Freight'] / freight_df_e['Quantity'])
freight_df_e = freight_df_e.sort_values(by='Dollars', ascending=False)
freight_df_e = freight_df_e[freight_df_e['Dollars'] > 100]
freight_df_e = freight_df_e[freight_df_e['Quantity'] <= 1000]
freight_df_e['FreightRate(%)'] = freight_df_e['FreightRate(%)'].round(2)
freight_df_e['FreightPerUnit'] = freight_df_e['FreightPerUnit'].round(2)
freight_df_e = freight_df_e[['Freight', 'Dollars', 'Quantity', 'FreightRate(%)', 'FreightPerUnit']]

freight_df_e.head(5)
```
