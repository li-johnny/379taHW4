```python
#find products with the highest margins on average
```

```python
sales = pd.read_csv("SalesFINAL12-31-16Sample.csv")
```

```python
purchasePrices = pd.read_csv("PurchasesFINAL12-31-16Sample.csv")
```

```python
merged_data = pd.merge(purchasePrices, sales, on='InventoryId')
```

```python
merged_data['ExciseTax'] = (merged_data['ExciseTax'] / merged_data['SalesQuantity'])
```

```python
merged_data['AvgMargin'] = merged_data['SalesPrice'] - merged_data['PurchasePrice'] - merged_data['ExciseTax']
```

```python
merged_data[['Description_x','AvgMargin', 'SalesPrice', 'PurchasePrice', 'ExciseTax']].groupby('Description_x').mean().sort_values(by='AvgMargin',ascending=False).head(10)
```

An internal accountant might use this calculation to recommend which brands to market more.
