# Deliverable 4c

```python
invoice_purchases['freight_per_$'] = invoice_purchases['Freight']/invoice_purchases['Dollars']
above_250000= invoice_purchases[invoice_purchases['Dollars']>250000]
above_250000.head()
```

```python
top_vendor_fpd = invoice_purchases[['VendorName','freight_per_$',]].groupby('freight_per_$').sum()
top_vendor_fpd.head(2)
```

```python
top_vendor_fpd.sort_values(by='freight_per_$',ascending=False).head(10)
```

```python
bottom_vendor_fpd = invoice_purchases[['VendorName','freight_per_$',]].groupby('freight_per_$').sum()
bottom_vendor_fpd.head(2)
```

```python
bottom_vendor_fpd.sort_values(by='freight_per_$',ascending=True).head(5)
```
