```python
dfLargeVendors = df2016Invoices.groupby(['VendorNumber','VendorName'])[['Freight','Dollars','Quantity']].sum()
```

```python
dfLargeVendors = dfLargeVendors[dfLargeVendors['Dollars']>=250000]
```

```python
dfLargeVendors.head()
```

```python
dfLargeVendors['FreightRate(%)'] = (dfLargeVendors['Freight']/dfLargeVendors['Dollars'])*100
dfLargeVendors['FreightPerUnit'] = (dfLargeVendors['Freight']/dfLargeVendors['Quantity'])
```

```python
dfLargeVendors = dfLargeVendors.sort_values(by='FreightRate(%)', ascending = False)
```

```python
dfLargeVendors.head(10)
```

```python
dfLargeVendors.sort_values(by='FreightRate(%)', ascending = True).head()
```

## These show the highest sellers and lowest sellers in December. This information is important because it helps management understand which products sell the best and which perform poorly which helps management make informed decisions about inventory purchases, pricing strategies, and marketing focus. For example, more resources can be allocated to promote high sellers to maximize profits, while strategies can be reconsidered for low sellers.
