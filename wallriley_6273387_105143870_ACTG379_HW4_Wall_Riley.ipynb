# Deliverable 5 (Finding Net income using Invoices and December Sales)

```python
dec_sales = pd.read_csv("salesDecember2016Sample.csv")
```

```python
print(dec_sales.columns)
```

```python
print(invoices.columns)
```

```python
merged_D5 = pd.merge(invoices, dec_sales, on=['InventoryId', 'Store', 'Brand', 'Description', 'Size', 
                                               'SalesQuantity', 'SalesDollars', 'SalesPrice', 'SalesDate', 
                                               'Volume', 'Classification', 'ExciseTax', 'VendorNo', 'VendorName'],
                       suffixes=('_invoices', '_dec_sales'))
```

```python
merged_D5['SalesDate'] = pd.to_datetime(merged_D5['SalesDate'])
```

```python
december_data = merged_D5[merged_D5['SalesDate'].dt.month == 12]
```

```python
total_cost_of_invoices = december_data['SalesDollars'].sum()
```

```python
total_revenue_for_december = december_data['SalesDollars'].sum()
```

```python
print("Total Cost of Invoices for December:", total_cost_of_invoices)
print("Total Revenue for December:", total_revenue_for_december)
```
