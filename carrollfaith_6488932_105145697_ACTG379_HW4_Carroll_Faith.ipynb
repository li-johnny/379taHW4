# Deliverable 4B

```python
vendors4B = invoicepurchases[['VendorName','VendorNumber','Freight']].groupby(
    ['VendorName','VendorNumber']).sum().sort_values(
    by= 'Freight', ascending=False)
vendors4B.head(10)
```

# Deliverable 4C

```python
vendors4C = invoicepurchases[['VendorName','VendorNumber','Quantity','Freight','Dollars']].groupby(
    ['VendorName','VendorNumber']).sum().sort_values(
    by= 'Freight', ascending=False)
vendors4C.head(10)
```

```python
vendors4C = vendors4C[vendors4C['Dollars']>= 250000]
vendors4C.head()
```

## Top 10 Vendors with the Highest per Dollar Freight

```python
vendors4C['FreightRate(%)'] = round((vendors4C['Freight'] / vendors4C['Dollars']) * 100, 2)
vendors4C['FreightperUnit'] = round(vendors4C['Freight'] / vendors4C['Quantity'],2)
vendors4C = vendors4C.sort_values(by= 'FreightRate(%)', ascending=False)
vendors4C.head(10)
```

## 5 Vendors with the Lowest per Dollar Freight

```python
vendors4C['FreightRate(%)'] = round((vendors4C['Freight'] / vendors4C['Dollars']) * 100, 2)
vendors4C['FreightperUnit'] = round(vendors4C['Freight'] / vendors4C['Quantity'],2)
vendors4C = vendors4C.sort_values(by= 'FreightRate(%)', ascending=True)
vendors4C.head(5)
```

# Deliverable 4D

# Deliverable 4E

```python
invoicepurchases['FreightRate(%)'] = round((invoicepurchases['Freight'] / invoicepurchases['Dollars']) * 100, 2)
invoicepurchases['FreightperUnit'] = round(invoicepurchases['Freight'] / invoicepurchases['Quantity'],2)
invoicepurchases.head()
```

```python
Smallex = invoicepurchases[(invoicepurchases['Freight'] > 100)]
Smallex = Smallex[(Smallex['Quantity'] <= 1000)]
Smallex.head()
```

```python
vendors4D['FreightRate(%)'] = round((vendors4D['Freight'] / vendors4D['Dollars']) * 100, 2)
vendors4D['FreightperUnit'] = round(vendors4D['Freight'] / vendors4D['Quantity'],2)
vendors4D.head()
```

```python
SmallExpensiveR = Smallex[['FreightRate(%)']].mean()
SmallExpensiveU = Smallex[['FreightperUnit']].mean()
LargeR = vendors4C[['FreightRate(%)']].mean()
LargeU = vendors4C[['FreightperUnit']].mean()
AllPurchasesR = invoicepurchases[['FreightRate(%)']].mean()
AllPurchasesU = invoicepurchases[['FreightperUnit']].mean()
```

```python
df = pd.DataFrame({
    "Small/Expensive": [SmallExpensiveR.values[0], SmallExpensiveU.values[0]],
    "LargeVendors": [LargeR.values[0], LargeU.values[0]],
    "AllPurchases": [AllPurchasesR.values[0], AllPurchasesU.values[0]]
})
df.style \
  .format(precision=3) \
  .format_index(str.upper, axis=1) \
  .relabel_index(["Freight%of$", "FreightperUnit"], axis=0)
```
