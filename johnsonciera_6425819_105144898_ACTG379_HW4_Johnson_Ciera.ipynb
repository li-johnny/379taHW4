## Deliverable 4a

```python
Purchases = pd.read_csv("PurchasesFINAL12312016.csv")
```

```python
Purchases_2016 = Purchases[(Purchases['InvoiceDate'] >= "2016-01-01") & (Purchases['InvoiceDate'] <= "2016-12-31")]
```

```python
Invoice_Purchases = pd.read_csv("InvoicePurchases12312016.csv")
```

```python
Top_Vendors_Dollars = Purchases_2016[['VendorName','Dollars']].groupby('VendorName').sum('Dollars')
```

```python
Top_Vendors_Dollars.sort_values(by='Dollars',ascending=False).head(10)
```

## Deliverable 4b

```python
Invoice_Purchases.head()
```

```python
Invoice_Purchases_2016 = Invoice_Purchases[(Invoice_Purchases['InvoiceDate'] >= "2016-01-01") & (Invoice_Purchases['InvoiceDate'] <= "2016-12-31")]
```

```python
Top_Vendors_Freight = Invoice_Purchases_2016[['VendorName','Freight', 'Quantity']].groupby('VendorName').sum('Freight')
```

```python
Top_Vendors_Freight.sort_values(by='Freight',ascending=False).head(10)
```

## Deliverable 4c

```python
Greater_250000 = Top_Vendors_Dollars[Top_Vendors_Dollars["Dollars"] >= 250000]
```

```python
Greater_250000.head()
```

```python
Merge_4c = pd.merge(left=Greater_250000,right=Top_Vendors_Freight, how='left', on=['VendorName'])
```

```python
Merge_4c["Freight per Dollar %"] = (Merge_4c["Freight"]/Merge_4c["Dollars"]*100) 
```

```python
Merge_4c.sort_values(by='Freight per Dollar %',ascending=False).head(10)
```

```python
Merge_4c.sort_values(by='Freight per Dollar %',ascending=False).tail(5)
```

## Deliverable 5

```python
Sales = pd.read_csv("SalesFINAL12312016.csv")
```

```python
Sales = Sales[['InventoryId', 'Description', 'Volume']].groupby('InventoryId').sum()
```

```python
Merge_5 = pd.merge(left=endInv[['InventoryId','InvCost12_31_2016']], right=Sales, how='outer', on=['InventoryId'])
```

Listed here are the ten products with the lowest volume per cost ratios. This information is useful because it can allow the company to see both how their inventory space stacks up to their spending, and which products have the least volume in sales per dollar spent on the inventory.
