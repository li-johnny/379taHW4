```python
begInv= pd.read_csv ("BegInvFINAL12-31-16Sample.csv")
```

```python
begInv_Store.sort_values(by='InvCost01_01_2016').tail(10).sort_values(by='InvCost01_01_2016', ascending = False)
```

```python
begInv_Brand = begInv[['Brand','01_01_2016_Qty','InvCost01_01_2016']].groupby('Brand').sum()
begInv_Brand.head(2)
```

```python
begInv_Brand.sort_values(by='InvCost01_01_2016').tail(10).sort_values(by='InvCost01_01_2016', ascending = False)
```

```python
begInv1["Beginning Total$"]= (begInv1["Price"]* begInv1["onHand"])

begInv1= begInv1.rename(columns = {'onHand':'Beginning Quantity'})
```

```python
begInv1.head()
```

```python
endInv1= pd.read_csv ('EndInvFINAL12-31-16Sample.csv')

endInv1["Ending Total$"]= (endInv1["Price"]* endInv1["onHand"])
endInv1= endInv1.rename(columns = {'onHand':'Ending Quantity'})
```

```python
endInv1.head()
```

```python
InventoryId= pd.merge(left= begInv1, right= endInv1, how='outer', 
on=['InventoryId'])
```

```python
InventoryId= InventoryId[["InventoryId", "Beginning Quantity", "Beginning Total$", "Ending Quantity", "Ending Total$"]]

InventoryId.fillna(0, inplace=True)
```

```python
Purchase = pd.read_csv("InvoicePurchases12-31-16Sample.csv")

Purchase_selected = Purchase[["VendorNumber", "VendorName", "Quantity", "Dollars", "Freight"]]
```

```python
Purchase_selected.head(10)
```

```python
Purchase_sum = Purchase_selected.set_index(['VendorNumber', 'VendorName']).groupby(['VendorNumber', 'VendorName']).sum()

Purchase_sum.head(2)
```

```python
Purchase_sum.sort_values(by='Dollars',ascending=False).head(10)
```

```python
Purchase_sum.sort_values(by='VendorNumber',ascending= True).head(10)
```

b.

```python
Purchase_sum.sort_values(by='Freight',ascending=False).head(10)
```

c.

```python
Purchase_sum["per $ freight"] = (Purchase_sum["Freight"] / Purchase_sum["Dollars"])

Purchase_sum["per $ freight"] = Purchase_sum["per $ freight"].round(5)

Purchase_sum['per $ freight'] = Purchase_sum['per $ freight'].apply(lambda x: '{:.5f}'.format(x))

Purchase_sum.head()
```

```python
above_250000 = Purchase_sum[Purchase_sum["Dollars"] > 250000]
```

```python
above_250000.head()
```

```python
above_250000.sort_values(by='per $ freight',ascending=False).head(10)
```

```python
above_250000.sort_values(by='per $ freight',ascending=True).head(5)
```

d.

```python
Purchase_sum_selected= Purchase_sum[(Purchase_sum['Freight']>100)| (Purchase_sum['Quantity'] <= 1000)]

Purchase_sum_selected.head(10)
```

```python
Purchase_sum_selected.sort_values(by='Freight',ascending=False).head(10)
```

```python
Purchase_from_d= Purchase_sum_selected.copy()

Purchase_from_d["per $ freight"] = (Purchase_from_d["Freight"] /Purchase_from_d["Dollars"])

Purchase_from_d["per $ freight"] = Purchase_from_d["per $ freight"].round(5)

Purchase_from_d['per $ freight'] = Purchase_from_d['per $ freight'].apply(lambda x: '{:.5f}'.format(x))


Purchase_from_d.head()
```
