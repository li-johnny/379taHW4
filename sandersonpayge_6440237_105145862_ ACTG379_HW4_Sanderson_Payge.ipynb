```python
#The purchases file was to large for Jupter to run, it would freeze on the kernel and stop running
#Bescaue of that I just used the InvoicePurchases file for all of Deliverable 4
Purchases = pd.read_csv(r"C:\Users\msand\Oregon State University\Pesch, Heather Lynn - Bibitor_Data\2016Data\SimplifiedData\HW4 Sampled Data\InvoicePurchases12312016.csv")
Purchases.head(2)
```

```python
Purchases['InvoiceDate'] = pd.to_datetime(Purchases["InvoiceDate"])
Purchases["year"] = Purchases["InvoiceDate"].dt.year
Purchases.head(2)
```

```python
Purchases2016 = Purchases[Purchases["year"] == 2016]
```

```python
Vendors2016 = Purchases2016[['VendorNumber','VendorName','Dollars']].groupby('VendorNumber').sum()
Vendors2016.sort_values(by='Dollars',ascending=False).head(10)
```

# Deliverable 4b

```python
Vendors_Freight = Purchases2016[['VendorNumber','VendorName','Freight', "Quantity"]].groupby('VendorNumber').sum()
Vendors_Freight.sort_values(by='Freight',ascending=False).head(10)
```

```python
#limit 4a to values above 250,000
#left merge that new table with 4b
#new colum with freight/dollars
#sortby the new column decening and find tail and head
```

### NOTE

Because I used the InvoicePurchses file, my tables are going to look a little different than the ones shown in the check figures

```python
High_Vendors = Vendors2016[Vendors2016["Dollars"] >= 250000]
High_Vendors_Freight = pd.merge(left = High_Vendors, right = Vendors_Freight, how="left", on=['VendorNumber', 'VendorName'])
High_Vendors_Freight.head(3)
```

```python
High_Vendors_Freight["Freight/$ %"] = (High_Vendors_Freight['Freight'] / High_Vendors_Freight['Dollars'])
High_Vendors_Freight["Freight/Unit"] = (High_Vendors_Freight['Freight'] / High_Vendors_Freight['Quantity'])
High_Vendors_Freight.head(2)
```

```python
High_Vendors_Freight['Freight/$ %'] = High_Vendors_Freight['Freight/$ %'].map('{:.2%}'.format)
```

```python
High_Vendors_Freight.sort_values(by='Freight/$ %', ascending=False).head(10)
```

```python
High_Vendors_Freight.sort_values(by='Freight/$ %', ascending=False).tail()
```

```python
#Which stores were the most profitable? Of those, what was their average inventory over the year? Could help look at inventory turnover. 
```

```python
Sales = pd.read_csv(r"C:\Users\msand\Oregon State University\Pesch, Heather Lynn - Bibitor_Data\2016Data\SimplifiedData\HW4 Sampled Data\SalesFINAL12312016Sample.csv")
Sales.head()
```

```python
TopStores = Sales[['Store','SalesDollars']].groupby('Store').sum()
TopStores.sort_values(by='SalesDollars',ascending=False).head()
```

```python
ABC = pd.merge(left=begInv_Store, right=endInv_Store, how="inner", on="Store")
ABC.head(2)
```

```python
ABC["AverageInv"] = ((ABC['InvCost01_01_2016'] + ABC['InvCost12_31_2016'])/2)
ABC.head(2)
```

```python
mid = pd.merge(left=TopStores, right=ABC[["AverageInv"]], how="inner", on="Store")
mid.head(2)
```

```python
mid["InvTurnover"] = (mid['SalesDollars']/mid["AverageInv"])
mid.head(2)
```

```python
mid['InvTurnover'] = mid['InvTurnover'].map('{:.2%}'.format)
mid.head(3)
```

```python
mid.sort_values(by='SalesDollars',ascending=False).head(10)
```
