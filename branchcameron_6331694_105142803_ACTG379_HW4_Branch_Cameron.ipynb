## Deliverable 5

In this anaylsis I will look at the cost of purchasing inventory from each vendor (not including freight) in 2016 compared to the revenue generated by the purchased inventory from each vendor in 2016. I will then calculate the amount of the surplus ordered from each vendor.

```python
sales = pd.read_csv("SalesFINAL12312016Sample.csv")
```

```python
sales.head()
```

```python
sales_2 = sales[['VendorName','SalesQuantity','SalesDollars']].groupby('VendorName').sum()
```

```python
sales_2.sort_values(by='SalesDollars',ascending=False).head(10)
```

```python
purchases.head()
```

```python
purchases_2 = purchases[['VendorName', 'Quantity','Dollars']].groupby('VendorName').sum()
```

```python
purchases_2.sort_values(by='Dollars',ascending=False).head(10)
```

```python
merge_2 = pd.merge(right = purchases_2, left = sales_2, on='VendorName', how='inner')
```

```python
merge_2 = merge_2.rename(columns = {'Dollars':'PurchaseCost'})
```

```python
merge_2 = merge_2.rename(columns = {'Quantity':'PurchaseQuantity'})
```

```python
merge_2 = merge_2.rename(columns = {'SalesDollars':'Revenue'})
```

```python
merge_2['2016VendorRevenue']= (merge_2['Revenue'] - merge_2['PurchaseCost'])
```

```python
merge_2['2016QuantitySurplus']= (merge_2['PurchaseQuantity'] - merge_2['SalesQuantity'])
```

```python
merge_2.sort_values(by='2016VendorRevenue',ascending=False).tail(10)
```
