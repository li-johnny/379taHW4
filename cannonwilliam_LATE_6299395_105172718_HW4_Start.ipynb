# Deliverable 4

```python
purchases = pd.read_csv("/Users/parkercannon/Desktop/HW #4/Bibitor_Data/2016Data/Full Dataset/InvoicePurchases12312016.csv")
purchases.head(2)
```

```python
purchases['InvoiceDate'] = pd.to_datetime(purchases['InvoiceDate'])
```

```python
purchases_2016 = purchases.loc[(purchases['InvoiceDate']>= '2016-01-01') & (purchases['InvoiceDate']<='2016-12-31')]
```

## Deliverable 4a

```python
best_vendors = purchases_2016[['VendorNumber','VendorName','Dollars']].groupby('VendorNumber').sum()
```

```python
best_vendors.sort_values(by='Dollars',ascending=False).head(10)
```

## Deliverable 4b

```python
freight_cost = purchases_2016[['VendorNumber','VendorName','Freight']].groupby('VendorNumber').sum()
```

```python
freight_cost.sort_values(by='Freight',ascending=False).head(10)
```

## Deliverable 4c

```python
purchases_4c = purchases_2016[purchases_2016['Dollars']>= 250000.00]
purchases_4c.head(10)
```

```python
purchases_4c['Freightper$%'] = ((purchases_4c['Freight'] / purchases_4c['Dollars'])*100)
purchases_4c['FreightperUnit'] = (purchases_4c['Freight']/purchases_4c['Quantity'])
```

```python
purchases_4c.head(2)
```

```python
top_purchases = purchases_4c[['VendorNumber','VendorName','Freightper$%','FreightperUnit']].groupby('VendorNumber').sum()
```

```python
top_purchases.sort_values(by='Freightper$%',ascending=False).head(10)
#Highest
```

```python
top_purchases.sort_values(by='Freightper$%',ascending=True).head(10)
#Lowest
```

## Deliverable 4d

```python
purchases_4d = purchases[(purchases["Freight"]>=100) & (purchases["Quantity"]<=1000)]
```

```python
purchases_4ds = purchases_4d[['VendorNumber','VendorName','Freight','Dollars','Quantity']].groupby(['VendorNumber','VendorName']).sum().sort_values(by='Freight',ascending=False)
```

```python
purchases_4ds.head(10)
```

## Deliverable 4e

```python
purchases_4ds['Freightper$%'] = ((purchases_4ds['Freight'] / purchases_4ds['Dollars'])*100)
purchases_4ds['FreightperUnit'] = (purchases_4ds['Freight']/purchases_4ds['Quantity'])
purchases_4ds.head(10)
```

```python
top_purchases.sort_values(by='Freightper$%',ascending=True).head(10)
#Lowest
```

### When comparing the chart from 4d and the chart from 4c, it is clear that the freight per percentage is more in line with the lowest values on the chart from 4c and the Freight per unit is much more inline with the lowest values on the chart from 4c

# Deliverable 5

```python
salesdata = pd.read_csv("/Users/parkercannon/Desktop/HW #4/Bibitor_Data/2016Data/Full Dataset/2017PurchasePricesDec.csv")
```

```python
salesdata.head(2)
```

```python
salesdata['Volume'] = pd.to_numeric(salesdata['Volume'],errors='coerce')
```

```python
salesdata['Volumeper$'] = (salesdata['Volume'] / salesdata['Price'])
```

```python
salesdata.head(2)
```

```python
volumeperdollar = salesdata[['Brand','Description','Size','Price','Volumeper$']]
```

```python
volumeperdollar.head(2)
```

```python
volumeperdollar = volumeperdollar.fillna(0)
```

```python
volumeperdollar.sort_values(by='Volumeper$',ascending = False).head(10)
```

## Displays how much volume of product per dollar the customer receives, its a statistic many customers want to see on a website when they are purchasing something of large volume
