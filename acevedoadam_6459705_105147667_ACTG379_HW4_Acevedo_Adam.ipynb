## Deliverable 1A

```python
begInv = pd.read_csv("BegInvFINAL12-31-16Sample.csv")
begInv = begInv.rename(columns = {'onHand':'01_01_2016_Qty'})
begInv['InvCost01_01_2016'] = begInv['01_01_2016_Qty'] * begInv['Price']
begInv_Store = begInv[['Store','01_01_2016_Qty','InvCost01_01_2016']].groupby('Store').sum()
begInv_Store.sort_values(by='InvCost01_01_2016',ascending=False).head(10)
begInv_Store.sort_values(by='InvCost01_01_2016').tail(10).sort_values(by='InvCost01_01_2016', ascending = False)
```

## Deliverable 4C

```python
InvoicePurchases = pd.read_csv("InvoicePurchases12-31-16Sample.csv")
InvoicePurchases["FreightPerDollar"] = InvoicePurchases["Freight"] / InvoicePurchases["Dollars"]
InvoicePurchases = InvoicePurchases[["VendorName", "Dollars", "Freight", "FreightPerDollar"]].groupby("VendorName").sum()
Top10Freight = InvoicePurchases[(InvoicePurchases["Dollars"] > 250000)]
Top10Freight.sort_values(by="Dollars", ascending=False).head(10)
```

```python
Top10Freight.sort_values(by="Dollars", ascending=False).tail()
```

## Deliverable 4D

```python
InvoicePurchases = pd.read_csv("InvoicePurchases12-31-16Sample.csv")
HighestFreight = InvoicePurchases[InvoicePurchases["Freight"] > 100]
HighestFreight = HighestFreight[HighestFreight["Quantity"] <= 1000]
HighestFreight = HighestFreight.groupby(["VendorNumber", "VendorName"])[["Freight", "Dollars", "Quantity"]].sum().sort_values(by= "Freight", ascending = False)
HighestFreight[["Freight"]].head()
```

## Deliverable 4E

```python
InvoicePurchases = pd.read_csv("InvoicePurchases12-31-16Sample.csv")
HighestFreight["Freight rate as %"] = (HighestFreight["Freight"]/HighestFreight["Dollars"] * 100)
HighestFreight["Freight Per Unit"] = (HighestFreight["Freight"] / HighestFreight["Quantity"] * 100)
InvoicePurchases["Freight rate as %"] = (InvoicePurchases["Freight"] / InvoicePurchases["Dollars"] * 100)
InvoicePurchases["Freight Per Unit"] = (InvoicePurchases["Freight"] / InvoicePurchases["Quantity"] * 100)
pd.DataFrame([["Freight Per Dollar", HighestFreight["Freight rate as %"].mean(), Top10Freight["FreightPerDollar"].mean(), 
               InvoicePurchases["Freight rate as %"].mean()], ["Freight Per Unit", HighestFreight["Freight Per Unit"].mean(), 
                Top10Freight["FreightPerDollar"].mean(), InvoicePurchases["Freight Per Unit"].mean()]], 
             columns = [" ", "Small", "Large", "Average"]).set_index(" ")
               
```

## Deliverable 5

```python
sales = pd.read_csv("SalesFINAL12-31-16Sample.csv")
purchases = pd.read_csv("InvoicePurchases12-31-16Sample.csv")
PurchaseTotal = purchases.groupby(["VendorName"])["Dollars"].sum()
SalesTotal = sales.groupby(["VendorName"])["SalesDollars"].sum()
TotalCM = pd.merge(left = sales, right = PurchaseTotal, how = "outer", on = ["VendorName"])
TotalCM["CM"] = (TotalCM["SalesDollars"] - TotalCM["Dollars"])
TotalCM = TotalCM[["VendorName", "SalesDollars", "Dollars", "CM"]]
TotalCM.head()
```
