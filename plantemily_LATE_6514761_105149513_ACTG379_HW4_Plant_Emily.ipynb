## Invoices

```python
Invoices = pd.read_csv("Invoices.csv")
Invoices["InvoiceDate"] = pd.to_datetime(Invoices["InvoiceDate"])
Invoices["Year"] = Invoices["InvoiceDate"].dt.year
Invoices = Invoices[Invoices["Year"] == 2016]
Invoices.head(2)
```

## Deliverable 4a

```python
TenVendInv = Invoices[["VendorNumber", "VendorName", "Dollars"]].groupby('VendorNumber').sum()
TenVendInv.sort_values(by='Dollars',ascending=False).head(10)
```

## Deliverable 4b

```python
TenVendFrt = Invoices[["VendorNumber", "VendorName", "Freight"]].groupby('VendorNumber').sum()
TenVendFrt.sort_values(by='Freight',ascending=False).head(10)
```

## Deliverable 4c

```python
Invoices1 = Invoices[["VendorNumber", "VendorName", "Freight", "Dollars", "Quantity"]].groupby(['VendorNumber',"VendorName"]).sum()
Invoices1["Frt%"] = ((Invoices1["Freight"] / Invoices1["Dollars"]) * 100)
Invoices1["FrtUnit"] = Invoices1["Freight"] / Invoices1["Quantity"]
```

```python
HighFrt = Invoices1
HighFrt = HighFrt[HighFrt["Dollars"] >= 250000]
HighFrt.sort_values(by='Frt%',ascending=False).head(10)
```

```python
LowFrt = Invoices1
LowFrt = LowFrt[LowFrt["Dollars"] >= 250000]
LowFrt.sort_values(by="Frt%").head(5)
```

## Deliverable 4d

```python
FrtCon = Invoices
FrtCon = FrtCon[FrtCon["Freight"] > 100]
FrtCon = FrtCon[FrtCon["Quantity"] <= 1000]
FrtCon = FrtCon[["VendorNumber", "VendorName", "Freight", "Dollars", "Quantity"]].groupby(['VendorNumber',"VendorName"]).sum()
FrtCon.sort_values(by="Freight",ascending=False).head()
```

## Deliverable 4e

```python
FrtCon["Frt%"] = ((FrtCon["Freight"] / FrtCon["Dollars"]) * 100)
FrtCon["FrtUnit"] = FrtCon["Freight"] / FrtCon["Quantity"]
FrtCon.sort_values(by="Frt%",ascending=False).head(10)
```

The average Freight % for the 10 highest Frt% vendors from 4c is 0.525, and the average Freight/Unit for the 10 highest Frt% vendors from 4c is 0.045.
The average Freight % for the 10 highest Frt% vendors from 4d is 3.838, and the average Freight/Unit for the 10 highest Frt% vendors from 4d is 0.446.
This shows that vendors with higher sales (above 250,000) generally have a much lower Freight % and Freight/Unit cost than vendors with lower sales (below 100), likely due to freight cost discounting that occurs as order sizes grow and shipments become more efficient.
