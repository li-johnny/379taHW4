# Deliverable 4C

Calculate freight per $ purchased for only those Vendors who Bibitor purchased at 
least $250,000 of inventory during 2016 and identify the ten Vendors with the 
highest per $ freight and the five Vendors with the lowest per $ freight

```python
df2016Inv_PurchasesG= df2016Inv_Purchases.groupby(['VendorName', 'VendorNumber'], as_index=False)[['Freight','Dollars','Quantity']].sum()
```

```python
df2016Inv_PurchasesG.head()
```

```python
dfLgVendors = df2016Inv_PurchasesG[df2016Inv_PurchasesG['Dollars']>250000]
dfLgVendors
```

```python
dfLgVendors['FreightRate(%)'] = (dfLgVendors['Freight']/dfLgVendors['Dollars'])*100
dfLgVendors['FreightPerUnit'] = (dfLgVendors['Freight']/dfLgVendors['Quantity'])
```

Ten Largest Vendors:

```python
dfLgVendors = dfLgVendors.sort_values(by='FreightRate(%)', ascending = False)
```

```python
dfLgVendors.head(10)
```

5 Smallest Vendors:

```python
dfLgVendors.sort_values(by ='FreightRate(%)', ascending = True).head()
```

```python
dfSmallExpensiveInvoices = df2016Inv_PurchasesG[df2016Inv_PurchasesG['Freight']>100]
dfSmallExpensiveInvoices
```
