# Deliverable 4c

```python
MajorVendors = Invoices.groupby(['VendorNumber', 'VendorName'])[['Freight', 'Dollars', 'Quantity']].sum()
```

```python
MajorVendors = MajorVendors[MajorVendors['Dollars'] >=250000]
```

```python
MajorVendors.head()
```

```python
MajorVendors['FreightRate(%)'] = (MajorVendors['Freight'] / MajorVendors['Dollars']) * 100 
```

```python
MajorVendors['FreightPerUnit'] = (MajorVendors['Freight'] / MajorVendors['Quantity'])
```

```python
MajorVendors = MajorVendors.sort_values(by='FreightRate(%)', ascending=False)
```

```python
MajorVendors.head(10)
```

```python
MajorVendors.sort_values(by= "FreightRate(%)", ascending = True).head()
```

# Deliverable 5

```python
SalesData = pd.read_csv('SalesFINAL12-31-16Sample.csv').groupby(["InventoryId"],as_index=False)[["SalesPrice"]].mean()
```

```python
SalesData.head()
```

```python
Purchases = pd.read_csv("PurchasesFINAL12-31-16Sample.csv").groupby(["InventoryId"],as_index=False)[["PurchasePrice"]].mean()
```

```python
Purchases.head()
```

```python
mergeA = pd.merge(left=SalesData, right=Purchases, how='outer', on=['InventoryId'])
```

```python
mergeA.head()
```

```python
mergeA['ContributionMargin'] = (mergeA['SalesPrice'] - mergeA['PurchasePrice'])
```

```python
mergeA.head()
```

```python
mergeA.sort_values(by='ContributionMargin', ascending = True).head(10)
```
