```python
purchases = pd.read_csv('PurchasesFINAL12312016.csv')
purchases['InvoiceDate'] = pd.to_datetime(purchases['InvoiceDate'])
purchases2016 = purchases[(purchases['InvoiceDate'] >= '2016-01-01') & (purchases['InvoiceDate'] <= '2016-12-31')]
purchases2016.head(10)
```

```python
len(purchases2016)
```

## Deliverable 4a

```python
top_vendors = purchases2016[['VendorNumber', 'VendorName', 'Dollars']].groupby('VendorNumber').sum()
top_vendors.sort_values(by = 'Dollars', ascending = False).head(10)
```

```python
invoices = pd.read_csv('InvoicePurchases12312016.csv')
invoices.head(10)
```

```python
vendor_freight = pd.merge(left = top_vendors, right = invoices, how = 'inner', on = 'VendorNumber')
vendor_freight.head(10)
```

## Deliverable 4b

```python
freight_dollars = vendor_freight[['VendorNumber', 'VendorName_y', 'Dollars_y', 'Freight']].groupby('VendorNumber').sum()
freight_dollars.sort_values(by = 'Freight', ascending = False).head(10)
```

## Deliverable 4c

```python
freight_dollars['Freight_Per_Dollar'] = (freight_dollars['Freight'] / freight_dollars['Dollars_y'])
freight_dollars.loc[freight_dollars['Dollars_y'] >= 250000].sort_values(by = 'Freight_Per_Dollar', ascending = False).head(10)
```

```python
freight_dollars.loc[freight_dollars['Dollars_y'] >= 250000].sort_values(by = 'Freight_Per_Dollar', ascending = False).tail(5)
```

## Deliverable 4d/e

```python
freight_dollars2 = vendor_freight[['VendorNumber', 'VendorName_y', 'Dollars_y', 'Freight', 'Quantity']].groupby('VendorNumber').sum()
freight_dollars2['Freight_Per_Dollar'] = (freight_dollars2['Freight'] / freight_dollars2['Dollars_y'])
freight_dollars2['Freight_Per_Unit'] = (freight_dollars2['Freight'] / freight_dollars2['Quantity'])
freight_dollars2.head()
```

```python
freight_dollars2 = freight_dollars2[(freight_dollars2['Freight'] > 100) & (freight_dollars2['Quantity'] <= 1000)]
freight_dollars2.head()
```
