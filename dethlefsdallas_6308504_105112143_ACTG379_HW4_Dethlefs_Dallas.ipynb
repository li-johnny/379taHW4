# Deliverable 4

```python
Purchases = pd.read_csv("PurchasesFINAL12-31-16Sample.csv")
Purchases.head()
```

```python
InvoicePurchases = pd.read_csv("InvoicePurchases12-31-16Sample.csv")
InvoicePurchases.head()
```

## Deliverable 4a

```python
TopVendors= Purchases[['VendorName', 'Dollars']].groupby('VendorName').sum()
TopVendors.sort_values(by='Dollars',ascending=False).head(10)
```

## Deliverable 4b

```python
TopFreight= InvoicePurchases[['VendorName', 'Freight']].groupby('VendorName').sum()
TopFreight.sort_values(by='Freight',ascending=False).head(10)
```

## Deliverable 4c

```python
Freightc= InvoicePurchases[['VendorName', 'Dollars', 'Freight']].groupby('VendorName').sum()
Freightc['Freightper$'] = Freightc['Freight'] / Freightc['Dollars']
Freightc250 = Freightc[Freightc["Dollars"] > 250000]
Freightc250.sort_values(by='Freightper$',ascending=False).head(10)
```

```python
Freightc250.sort_values(by='Freight',ascending=True).head(10)
```

## Deliverable 4d

```python
Freightd= InvoicePurchases[['VendorName', 'Dollars', 'Freight', 'Quantity']]
Freightd100= Freightd[Freightd["Freight"] > 100]
Freightd1000= Freightd100[Freightd100["Quantity"] <= 1000]
Freightd1000.groupby('VendorName').sum().sort_values(by='Freight',ascending=False).head(10)
```

The vendors with the highest freight costs under these conditions include Edrington Americas, Stoli Group(USA) LLC, and Remy Cointreau USA Inc. The table shows the rest of the top 10

## Deliverable 4e

```python
Freighte= Freightd1000[['VendorName', 'Dollars', 'Freight', 'Quantity']]
Freighte['Freightper$'] = Freightd1000['Freight'] / Freightd1000['Dollars']
Freighte['Freightperunit'] = Freightd1000['Freight'] / Freightd1000['Quantity']
Freighte.sort_values(by='Freight',ascending=False).head(10)
```

```python
Freighte['Freightper$'].mean(axis=0,skipna=True,numeric_only=True)
```

The average freightper$ from part e is much higher than the results from part c

# Deliverable 5

```python
SalesDecember = pd.read_csv("salesDecember2016Sample.csv")
SalesDecember.head(2)
```

```python
SalesDecembert= SalesDecember[['InventoryId', 'Description', 'SalesQuantity', 'SalesDollars']].groupby(['InventoryId', 'Description']).sum().sort_values(by='SalesQuantity', ascending=False)
SalesDecembert.head()
```

```python
TopPurchases= Purchases[['InventoryId', 'Description', 'Quantity', 'Dollars']].groupby(['InventoryId', 'Description']).sum().sort_values(by='Quantity', ascending=False)
TopPurchases.head()
```

```python
Merge5= pd.merge(left=TopPurchases, right=SalesDecembert, how='outer',on=['InventoryId'])
Merge5.sort_values(by='SalesQuantity', ascending=False).head()
```

The table above shows the top selling items in December in comparison to the quantity of that item purchased during the year as well as salesdollars in december compared to total dollars spent on purchases during that year for each item
