# Deliverable 4c

```python
dfLargeVendors = freight.groupby(['VendorNumber', 'VendorName'])[['Freight','Dollars','Quantity']].sum()
```

```python
dfLargeVendors.head()
```

```python
dfLargeVendors = dfLargeVendors[dfLargeVendors['Dollars']>=250000]
```

```python
dfLargeVendors.head(10)
```

```python
dfLargeVendors['FreightRate(%)'] = (dfLargeVendors['Freight']/dfLargeVendors['Dollars'])*100
dfLargeVendors['FreightPerUnit'] = (dfLargeVendors['Freight']/dfLargeVendors['Quantity'])
```

```python
dfLargeVendors = dfLargeVendors.sort_values(by='FreightRate(%)', ascending = False)
```

```python
dfLargeVendors.sort_values(by='FreightRate(%)', ascending = True).head()
```

# Deliverable 5 

```python
salesdata = pd.read_csv("SalesFINAL12-31-16Sample.csv").groupby(["InventoryId"],as_index=False)[["SalesPrice"]].mean()
```

```python
salesdata.head()
```

```python
purchases = pd.read_csv("PurchasesFINAL12-31-16Sample.csv").groupby(["InventoryId"],as_index=False)[["PurchasePrice"]].mean()
```

```python
purchases.head()
```

```python
mergeA = pd.merge(left=salesdata, right=purchases, how='outer', on=['InventoryId'])
```

```python
mergeA.head()
```

```python
mergeA['Contribution Margin'] = (mergeA['SalesPrice'] - mergeA['PurchasePrice'])
```

```python
mergeA.head()
```

```python
mergeA = mergeA[mergeA["Contribution Margin"]<=0]
mergeA.sort_values(by='Contribution Margin', ascending = True)
```
