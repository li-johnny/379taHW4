# 5
Using the sales data and at least some of the other data, do an analysis of your choice.

```python
dfSalesPrice = pd.read_csv("SalesFINAL12-31-16Sample.csv").groupby(['Description'])[['SalesPrice']].mean()
dfPurchasePrice = dfPurchases.groupby(['Description'])[['PurchasePrice']].mean()
```

```python
dfCM = pd.merge(left = dfSalesPrice, right = dfPurchasePrice, how='inner', on='Description')
dfCM['Contribution Margin'] = dfCM['SalesPrice'] - dfCM['PurchasePrice']
```

```python
dfCM = dfCM[dfCM["Contribution Margin"]<=0]
dfCM.sort_values(by='Contribution Margin', ascending = True)
```

There are many different ways to analyze this data. The above code calculates the average contribution margin per unit and finds all of the brands that the company sold at a loss. <br>
An internal accountant might use this calculation to recommend which brands to discontinue. <br>
An auditor might use this to test inventory valuation at year end. Brands on this list remaining in inventory at year end might need to be adjusted to the lower of cost or market (LCM).
